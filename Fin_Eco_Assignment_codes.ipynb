```python
# Granger causality test
max_lag = 10
granger_test = grangercausalitytests(data[['Futures_Price', 'Spot_Price']].dropna(), max_lag, verbose=True)
```

```python
# Johansen Co-integration Test
def johansen_test(df, column1, column2):
    coint_test = coint_johansen(df[[column1, column2]], det_order=0, k_ar_diff=1)
    return coint_test.lr1, coint_test.cvt

coint_test_stat, coint_critical_values = johansen_test(data, 'Futures_Price', 'Spot_Price')
print(f'Co-integration Test Statistic: {coint_test_stat}')
print(f'Critical Values: {coint_critical_values}')
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Assuming 'data' is your DataFrame with 'Futures_Price' and 'Spot_Price'

# Step 1: Calculate the residuals (ECT) from the co-integration relationship
# Let's say Spot_Price is the dependent variable and Futures_Price is the independent variable
data['residual'] = data['Spot_Price'] - sm.OLS(data['Spot_Price'], sm.add_constant(data['Futures_Price'])).fit().fittedvalues

# Step 2: Calculate the first differences
data['dSpot_Price'] = data['Spot_Price'].diff()
data['dFutures_Price'] = data['Futures_Price'].diff()

# Drop NA values generated by differencing
data.dropna(inplace=True)

# Step 3: Check for and handle any remaining inf or NaN values
# This will print the rows with any inf or NaN values
print(data[~data.replace([np.inf, -np.inf], np.nan).dropna().index])

# Remove rows with inf or NaN values
data = data.replace([np.inf, -np.inf], np.nan).dropna()

# Step 4: Estimate the ECM
ecm_model = sm.OLS(data['dSpot_Price'], sm.add_constant(np.column_stack((data['dFutures_Price'], data['residual'].shift(1))))).fit()

# Step 5: Summary of ECM
print(ecm_model.summary())

```

```python
import pandas as pd

# Assuming you have a DataFrame 'data' with columns 'Date', 'Spot Price', 'Futures Price'
# and a carrying cost rate (e.g., 2% per annum)

# Convert annual carrying cost rate to daily rate (assume 252 trading days in a year)
carrying_cost_rate = 0.02 / 252

# Calculate carrying costs
data['Carrying Cost'] = data['Spot_Price'] * carrying_cost_rate

# Calculate the arbitrage condition
data['Arbitrage Opportunity'] = data['Futures_Price'] - (data['Spot_Price'] + data['Carrying Cost'])

# Identify where arbitrage is possible
data['Arbitrage Signal'] = data['Arbitrage Opportunity'] > 0

# View results
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Plot Spot and Futures Prices
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Spot_Price'], label='Spot_Price')
plt.plot(data['Date'], data['Futures_Price'], label='Futures_Price')
plt.title('Spot vs Futures Prices')
plt.xlabel('Year_month')
plt.ylabel('Price(₹/kg)')
plt.legend()
plt.savefig('fig_02.png')
plt.show()

# Calculate Moving Averages
data['Spot_MA'] = data['Spot_Price'].rolling(window=20).mean()
data['Futures_MA'] = data['Futures_Price'].rolling(window=20).mean()

# Plot Moving Averages
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Spot_Price'], label='Spot_Price')
plt.plot(data['Date'], data['Spot_MA'], label='Spot Price MA')
plt.plot(data['Date'], data['Futures_Price'], label='Futures Price')
plt.plot(data['Date'], data['Futures_MA'], label='Futures Price MA')
plt.title('Spot vs Futures Prices with Moving Averages')
plt.xlabel('Year_month')
plt.ylabel('Price(₹/kg)')
plt.legend()
plt.savefig('fig_03.png')
plt.show()
```

```python
# Calculate correlation between spot and futures prices
correlation = data['Spot_Price'].corr(data['Futures_Price'])

print(f"Correlation between Spot Price and Futures Price: {correlation:.2f}")
```

```python
# Calculate daily returns for spot and futures prices
data['Spot_Return'] = data['Spot_Price'].pct_change()
data['Futures_Return'] = data['Futures_Price'].pct_change()

# Calculate rolling volatility (standard deviation) over a window (e.g., 20 days)
data['Spot_Volatility'] = data['Spot_Return'].rolling(window=20).std()
data['Futures_Volatility'] = data['Futures_Return'].rolling(window=20).std()

# Plot Volatility
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Spot_Volatility'], label='Spot Volatility')
plt.plot(data['Date'], data['Futures_Volatility'], label='Futures Volatility')
plt.title('Spot vs Futures Volatility')
plt.xlabel('Year_month')
plt.ylabel('Volatility')
plt.legend()
plt.savefig('fig_04.png')
plt.show()
```
